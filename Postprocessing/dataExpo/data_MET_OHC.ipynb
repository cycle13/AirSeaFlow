```python
%matplotlib inline

import sys
sys.path.append("/home/yang/NLeSC/Computation_Modeling/Bjerknes/Scripts/META")
import numpy as np
import scipy
from netCDF4 import Dataset
import matplotlib
import matplotlib.pyplot as plt
import os
import meta.statistics
import meta.visualizer
```

Initialization - Start with location of input and extraction of variables
Time span of each product:
- **ERA-Interim** 1979 - 2016
- **MERRA2**      1980 - 2016
- **JRA55**       1979 - 2015
- **ORAS4**       1958 - 2014
- **GLORYS2V3**   1993 - 2014
- **SODA3**       1980 - 2015

```python
if __name__=="__main__":
    # sample
    ################################   Input zone  ######################################
    # specify data path
    datapath = '/home/yang/workbench/Core_Database_AMET_OMET_reanalysis/warehouse_met'
    datapath_ERAI_fields = '/home/yang/workbench/Core_Database_AMET_OMET_reanalysis/ERAI/regression'
    output_path = '/home/yang/NLeSC/Computation_Modeling/Bjerknes/Compensation/dataExpo'
    #####################################################################################
    print ('*********************** extract variables *************************')
    dataset_ERAI_fields_SIC_SST_SLP = Dataset(os.path.join(datapath_ERAI_fields,
                                              'surface_ERAI_monthly_regress_1979_2016.nc'))
    dataset_ERAI_AMET = Dataset(os.path.join(datapath,
                                             'era_model_subdaily_1979_2016_amet.nc'))
    dataset_ORAS4_OMET = Dataset(os.path.join(datapath,
                                              'oras_model_monthly_1958_2017_omet.nc'))
    dataset_ORAS4_OHC = Dataset(os.path.join(datapath,
                                              'oras_model_monthly_1958_2017_ohc.nc'))
    # extract time series from 1979 to 2016
    # from 20N - 90N
    # MET
    # 3D fields [year, month, lat]
    # since OMET is from 20N - 90N, AMET is from 90N to 20N, we have to reverse it
    # for interpolation, x should be monotonically increasing
    AMET_0_ERAI = dataset_ERAI_AMET.variables['E_total_int'][:,:,::-1]/1000 # from Tera Watt to Peta Watt
    AMET_200_ERAI = dataset_ERAI_AMET.variables['E_200_int'][:,:,::-1]/1000
    AMET_500_ERAI = dataset_ERAI_AMET.variables['E_500_int'][:,:,::-1]/1000
    AMET_850_ERAI = dataset_ERAI_AMET.variables['E_850_int'][:,:,::-1]/1000
    # 4D fields [year, month, level, lat]
    #AMET_vert_ERAI_reverse = dataset_ERAI_AMET.variables['E_vert'][:]/1000
    # 3D fields [year, month, lat]
    OMET_int_ORAS4 = dataset_ORAS4_OMET.variables['E_int'][21:-1,:,180:]/1000 # from Tera Watt to Peta Watt # start from 1979
    OMET_atl_int_ORAS4 = dataset_ORAS4_OMET.variables['E_atl_int'][21:-1,:,180:]/1000
    OMET_atl_100_int_ORAS4 = dataset_ORAS4_OMET.variables['E_atl_100_int'][21:-1,:,180:]/1000
    OMET_atl_300_int_ORAS4 = dataset_ORAS4_OMET.variables['E_atl_300_int'][21:-1,:,180:]/1000
    OMET_atl_700_int_ORAS4 = dataset_ORAS4_OMET.variables['E_atl_700_int'][21:-1,:,180:]/1000
    OMET_atl_2000_int_ORAS4 = dataset_ORAS4_OMET.variables['E_atl_2000_int'][21:-1,:,180:]/1000
    
    OMET_pac_int_ORAS4 = dataset_ORAS4_OMET.variables['E_pac_int'][21:-1,:,180:]/1000
    # 3D fields [year, month, lat]
    OHC_int_ORAS4 = dataset_ORAS4_OHC.variables['OHC_int'][21:-1,:,180:]/1000 # from Tera Watt to Peta Watt # start from 1979
    OHC_atl_int_ORAS4 = dataset_ORAS4_OHC.variables['OHC_atl_int'][21:-1,:,180:]/1000
    OHC_atl_100_int_ORAS4 = dataset_ORAS4_OHC.variables['OHC_atl_100_int'][21:-1,:,180:]/1000
    OHC_atl_300_int_ORAS4 = dataset_ORAS4_OHC.variables['OHC_atl_300_int'][21:-1,:,180:]/1000
    OHC_atl_700_int_ORAS4 = dataset_ORAS4_OHC.variables['OHC_atl_700_int'][21:-1,:,180:]/1000
    OHC_atl_2000_int_ORAS4 = dataset_ORAS4_OHC.variables['OHC_atl_2000_int'][21:-1,:,180:]/1000   
    # other dimensions
    year_ORAS4 = dataset_ORAS4_OMET.variables['year'][21:-1]    # from 1979 to 2016
    year_ERAI = dataset_ERAI_AMET.variables['year']
    latitude_OMET_ORAS4 = dataset_ORAS4_OMET.variables['latitude_aux'][180:]
    latitude_OHC_ORAS4 = dataset_ORAS4_OHC.variables['latitude_aux'][180:]
    latitude_AMET_ERAI = dataset_ERAI_AMET.variables['latitude'][::-1]
```

```python
# Get the latitudes of interest
print ('*********************** latitudes of interest *************************')
# AMET ERAI
latitude_AMET_ERAI_60 = latitude_AMET_ERAI[54]
#print (latitude_AMET_ERAI_60)
latitude_AMET_ERAI_55 = latitude_AMET_ERAI[47]
#print (latitude_AMET_ERAI_55)
latitude_AMET_ERAI_50 = latitude_AMET_ERAI[41]
#print (latitude_AMET_ERAI_50)
# OMET ORAS4
latitude_OMET_ORAS4_60 = latitude_OMET_ORAS4[53]
#print (latitude_OMET_ORAS4_60)
latitude_OMET_ORAS4_55 = latitude_OMET_ORAS4[45]
#print (latitude_OMET_ORAS4_55)
latitude_OMET_ORAS4_50 = latitude_OMET_ORAS4[38]
#print (latitude_OMET_ORAS4_50)
# OHC ORAS4
latitude_OHC_ORAS4_60 = latitude_OHC_ORAS4[53]
latitude_OHC_ORAS4_55 = latitude_OHC_ORAS4[45]
latitude_OHC_ORAS4_50 = latitude_OHC_ORAS4[38]
print (latitude_OHC_ORAS4_50)
```

```python
print ('*********************** post-processing *************************')
lowpass_window = 60

stat_AMET_ERAI_0 = meta.statistics.operator(AMET_0_ERAI)
stat_AMET_ERAI_0.anomaly()
stat_AMET_ERAI_0.detrend()
stat_AMET_ERAI_0.lowpass(obj='detrend')
stat_AMET_ERAI_200 = meta.statistics.operator(AMET_200_ERAI)
stat_AMET_ERAI_200.anomaly()
stat_AMET_ERAI_200.detrend()
stat_AMET_ERAI_200.lowpass(obj='detrend')
stat_AMET_ERAI_500 = meta.statistics.operator(AMET_500_ERAI)
stat_AMET_ERAI_500.anomaly()
stat_AMET_ERAI_500.detrend()
stat_AMET_ERAI_500.lowpass(obj='detrend')
stat_AMET_ERAI_850 = meta.statistics.operator(AMET_850_ERAI)
stat_AMET_ERAI_850.anomaly()
stat_AMET_ERAI_850.detrend()
stat_AMET_ERAI_850.lowpass(obj='detrend')

stat_OMET_ORAS4_atl_0 = meta.statistics.operator(OMET_atl_int_ORAS4)
stat_OMET_ORAS4_atl_0.anomaly()
stat_OMET_ORAS4_atl_0.detrend()
stat_OMET_ORAS4_atl_0.lowpass(obj='detrend')
stat_OMET_ORAS4_atl_100 = meta.statistics.operator(OMET_atl_100_int_ORAS4)
stat_OMET_ORAS4_atl_100.anomaly()
stat_OMET_ORAS4_atl_100.detrend()
stat_OMET_ORAS4_atl_100.lowpass(obj='detrend')
stat_OMET_ORAS4_atl_300 = meta.statistics.operator(OMET_atl_300_int_ORAS4)
stat_OMET_ORAS4_atl_300.anomaly()
stat_OMET_ORAS4_atl_300.detrend()
stat_OMET_ORAS4_atl_300.lowpass(obj='detrend')
stat_OMET_ORAS4_atl_700 = meta.statistics.operator(OMET_atl_700_int_ORAS4)
stat_OMET_ORAS4_atl_700.anomaly()
stat_OMET_ORAS4_atl_700.detrend()
stat_OMET_ORAS4_atl_700.lowpass(obj='detrend')
stat_OMET_ORAS4_atl_2000 = meta.statistics.operator(OMET_atl_2000_int_ORAS4)
stat_OMET_ORAS4_atl_2000.anomaly()
stat_OMET_ORAS4_atl_2000.detrend()
stat_OMET_ORAS4_atl_2000.lowpass(obj='detrend')

stat_OHC_ORAS4_atl_0 = meta.statistics.operator(OHC_atl_int_ORAS4)
stat_OHC_ORAS4_atl_0.anomaly()
stat_OHC_ORAS4_atl_0.detrend()
stat_OHC_ORAS4_atl_0.lowpass(obj='detrend')
stat_OHC_ORAS4_atl_100 = meta.statistics.operator(OHC_atl_100_int_ORAS4)
stat_OHC_ORAS4_atl_100.anomaly()
stat_OHC_ORAS4_atl_100.detrend()
stat_OHC_ORAS4_atl_100.lowpass(obj='detrend')
stat_OHC_ORAS4_atl_300 = meta.statistics.operator(OHC_atl_300_int_ORAS4)
stat_OHC_ORAS4_atl_300.anomaly()
stat_OHC_ORAS4_atl_300.detrend()
stat_OHC_ORAS4_atl_300.lowpass(obj='detrend')
stat_OHC_ORAS4_atl_700 = meta.statistics.operator(OHC_atl_700_int_ORAS4)
stat_OHC_ORAS4_atl_700.anomaly()
stat_OHC_ORAS4_atl_700.detrend()
stat_OHC_ORAS4_atl_700.lowpass(obj='detrend')
stat_OHC_ORAS4_atl_2000 = meta.statistics.operator(OHC_atl_2000_int_ORAS4)
stat_OHC_ORAS4_atl_2000.anomaly()
stat_OHC_ORAS4_atl_2000.detrend()
stat_OHC_ORAS4_atl_2000.lowpass(obj='detrend')
```

```python
print ('*********************** plots factory *************************')
index_year = np.arange(1985,year_ERAI[-1]+1,5)
index_x = np.arange(1, len(year_ERAI)*12-(lowpass_window-1)+1,1)

# AMET ERA-Interim at 60N
fig1 = plt.figure()
plt.plot(index_x, stat_AMET_ERAI_0._lowpass[:,54], 'b-', label='Total')
plt.plot(index_x, stat_AMET_ERAI_200._lowpass[:,54], 'r-', label='200hPa')
plt.plot(index_x, stat_AMET_ERAI_500._lowpass[:,54], 'g-', label='500hPa')
plt.plot(index_x, stat_AMET_ERAI_850._lowpass[:,54], 'm-', label='850hPa')
fig1.set_size_inches(12.5, 6)
plt.xlabel("Time", fontsize=16)
plt.xticks(np.arange(13,len(year_ERAI)*12+1,60), index_year, fontsize=16)
plt.ylabel("Meridional Energy Transport (PW)",fontsize=16)
plt.yticks(fontsize=16)
plt.legend(frameon=True, loc=2, prop={'size': 16})
plt.show()
fig1.savefig(os.path.join(output_path, 'data_ERAI_amet_series_60N.png'), dpi=300)
```
